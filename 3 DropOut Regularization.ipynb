# DropOut Regularization

```python
import pandas as pd
```

```python
data = pd.read_csv("data/sonar data.csv", header=None)
```

```python
data
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data[60].value_counts()
```

```python
X = data.drop(60, axis=1)
y = data[60]
```

```python
y
```

```python
axe = {"R" : 1, "M" : 0}
y = y.map(axe)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=51, stratify=y)
```

```python
X_train.shape, X_test.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

Overfitted Model

```python
model = keras.Sequential([
    layers.Dense(60, input_shape=(60,), activation="relu"),
    layers.Dense(30, activation="relu"),
    layers.Dense(15, activation="relu"),
    layers.Dense(1, activation="sigmoid"),
])
```

```python
model.compile(loss="binary_crossentropy", optimizer="rmsprop", metrics=["accuracy", keras.metrics.AUC()])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

Regularized Model

```python
model = keras.Sequential([
    layers.Dense(64, input_shape=(60,), activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(32, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(64, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(1, activation="sigmoid"),
])
```

```python
model.compile(loss="binary_crossentropy", optimizer="rmsprop", metrics=["accuracy", keras.metrics.AUC()])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```
